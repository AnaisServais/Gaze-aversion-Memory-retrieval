```python
###################################
###################################
# EXTRACTION DES DATAS OPENSESAME #
###################################
###################################
import csv
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
import cv2 
import pandas as pd

# change path directory
set_WD = os.getcwd() # get current work directory

idd={'1':'P01','2':'P02','3':'P03','4':'P04','5':'P05', \
     '6':'P06','7':'P07','9':'P09','10':'P10','11':'P11', \
     '12':'P12','13':'P13','14':'P14','15':'P15','16':'P16', \
     '17':'P17','18':'P18', '19': 'P19', '20': 'P20', '21': 'P21', '22': 'P22', '23':'P23', '24':'P24', '25':'P25', '26':'P26', '27':'P27', \
     '28': 'P28', '29':'P29', '30':'P30', '31':'P31', '32':'P32', '33':'P33'} #list of all the participants

wd_outpout=set_WD+"/data/processed_data/T1_visu.csv"
wd_input=set_WD+"/data/raw_data/task"
```

```python
def ouvrir_fichier() :
    f=open(wd_outpout, "a")
    L="Sujet,sujet_num,Bloc,audio,Question,end_acces,start_visu,clarte,tempau,effort \n"
    f.writelines(L)
    f.close()
def completer_fichier(suj,suj_num,bloc,aud,numero,rep,tps,clarte,tempau,effort) :
    f=open(wd_outpout, "a")
    FF=suj+','+str(suj_num)+','+str(bloc)+','+aud+','+numero+','+str(rep)+','+str(tps)+','+clarte+','+tempau+','+effort+'\n'
    f.writelines(FF)
    f.close()
        
ouvrir_fichier()
os.chdir(wd_input)
lst = os.listdir()
lst.sort()

for filename in lst :
    num=1
    data = pd.read_csv(filename,sep=',')
    sujet=filename[8:-6]
    bloc=filename[-5:-4]
    audio=data['audio']
    if float(sujet)<7:
        reponse=data['response_time_fin_acces']
        time=data['response_time_Debriefing']
    elif float(sujet)>6:
        reponse=data['response_time_fin_acces']
        time=data['response_time_Debriefing']
    clarte=data['response_Eval_clarte']
    tempau=data['response_Eval_tempau']
    effort=data['response_Eval_effort']    
    for i in range(len(data)) :
        aud=audio[i]
        rep=reponse[i]
        tps=str(time[i])
        clart=str(clarte[i])
        temp=str(tempau[i])
        eff=str(effort[i])
        numero=str(num)
        completer_fichier(idd[sujet],sujet,bloc,aud,numero,rep,tps,clart,temp,eff)
        num=num+1
 
```

```python
##############################
##############################
# EXTRACTION QUALITY DEBRIEF #
##############################
##############################


wd_outpout=set_WD+"/data/processed_data/quality_visu.csv"
wd_input=set_WD+'/data/raw_data/richesse_debrief'

def ouvrir_fichier() :
    f=open(wd_outpout, "a")
    L="Sujet,Bloc,Question,Etape_1,Etape_2\n"
    f.writelines(L)
    f.close()
def completer_fichier(suj,bloc,numero,etape_1,etape_2) :
    f=open(wd_outpout, "a")
    FF=suj+','+bloc+','+numero+','+etape_1+','+etape_2+'\n'
    f.writelines(FF)
    f.close()
        
ouvrir_fichier()
os.chdir(wd_input)
lst = [x for x in os.listdir() if x[-3:]=="csv"]
lst.sort()

for filename in lst : 
    print ("reading the file :",filename)
    data = pd.read_csv(filename,sep=',')   
    num=data['Question']
    sujet=data['Sujet']
    blocc=data['Bloc']
    step_1=data['Etape 1 result']
    step_2=data['Etape 2 result']    
    for i in range(len(data)) :
        numero=str(num[i])
        suj=str(sujet[i])
        bloc=str(blocc[i])
        etape_2=str(step_2[i])
        etape_1=step_1[i]
        completer_fichier(suj,bloc,numero,etape_1,etape_2)
 
```

```python
def convert_to_str(input_seq, seperator):
   final_str = seperator.join(input_seq)
   return final_str
```

```python
############################
# FUSION E.T. & T1+QUALITY #
############################

file=set_WD+"/data/processed_data/output_visu.csv"
a = pd.read_csv(set_WD+"/data/processed_data/ratios_visu.csv")
b = pd.read_csv(set_WD+"/data/processed_data/T1_quality_visu.csv")
merged = b.merge(a)
merged.to_csv(file, index=False)

f=open(set_WD+"/data/processed_data/data_T1_visu.csv", "a")
inp=csv.reader(open(file,"r"))

ind_ligne=0
for row in inp:  
    if row[15]!='MISSING' and row[10]!='NoA':        
        L=convert_to_str(row, ',')
        #print(L)
        if ind_ligne==1:
            if float(row[12])==0:
                L=L+',0'
            elif float(row[12])>0:
                L=L+',1'            
        if ind_ligne==0:
            L=L+',VD_here'
            ind_ligne=1
        L=L+'\n'
        f=open(set_WD+"/data/processed_data/data_T1_visu.csv", "a")
        f.writelines(L)
        f.close()
```

# Open data_T1_visu.csv 
To adjust the particular case where a VD started in access and continues in elaboration 
We want to count them as VD during elaboration and to take them into account in the calculation of the ratio, duration, etc.

```python
df_data_T1_visu = pd.read_csv(set_WD+"/data/processed_data/data_T1_visu.csv")
path_output = set_WD+"/data/processed_data/data_T1_visu.csv" # the data from the VD starting in access and continuing in elaboration are added in the descriptive statistics values

df_data_T1_visu
```

```python
for i in range(len(df_data_T1_visu['Sujet'])):
    if df_data_T1_visu['delai'][i] == 'START_during_ACCESS':
        df_data_T1_visu['nb_VD'][i] = df_data_T1_visu['nb_VD'][i] + 1
df_data_T1_visu
```

```python
for i in range(len(df_data_T1_visu['Sujet'])):
    if df_data_T1_visu['delai'][i] == 'START_during_ACCESS':
        df_data_T1_visu['moy_duree_VD'][i] = df_data_T1_visu['moy_duree_VD'][i] + df_data_T1_visu['duration_VD_in_elaboration_if_started_before '][i]/df_data_T1_visu['nb_VD'][i]
df_data_T1_visu
```

```python
duration_elaboration = 10000 # 10 sec
for i in range(len(df_data_T1_visu['Sujet'])):
    if df_data_T1_visu['delai'][i] == 'START_during_ACCESS':
        df_data_T1_visu['ratio'][i] = df_data_T1_visu['ratio'][i] + df_data_T1_visu['duration_VD_in_elaboration_if_started_before '][i]/duration_elaboration
df_data_T1_visu
```

### To remove questions where the VD started before memory retrieval

```python
df_data_T1_visu
df_data_T1 = pd.read_csv(set_WD+"/data/processed_data/data_T1_access.csv")
```

```python
ind_nb = []
for i in range(len(df_data_T1_visu['Sujet'])):
    
    suj_T1 = df_data_T1[df_data_T1['Sujet'] == df_data_T1_visu['Sujet'][i]]
    suj_T1 = suj_T1[suj_T1['Bloc'] == df_data_T1_visu['Bloc'][i]]
    
    if df_data_T1_visu['audio'][i] not in list(suj_T1['audio']):
        
        ind_nb.append(i)


for ind in ind_nb:
    
    df_data_T1_visu = df_data_T1_visu.drop(ind)
df_data_T1_visu
```

```python
df_data_T1_visu.reset_index(inplace = True)
df_data_T1_visu
```
